### 1. Imports and Dataset Analysis


The following are achieved - 
 - Create seperate lists for each UN Region
 - Create combined dataset with data from year 1961 to 2018

```python
import pandas as pd

# Defining filepath
filepath_T2014 = 'Data/Till2014.csv'
filepath_2015 = 'Data/2015.csv'
filepath_2016 = 'Data/2016.csv'
filepath_2017 = 'Data/2017.csv'
filepath_2018 = 'Data/2018.csv'

# Load file with `sheet_name=None` - returns a dictionary
# Original dataset from Kaggle with data until 2014
df_T2014 = pd.read_csv(filepath_T2014, encoding = 'unicode_escape')

# Data obtained from API calls through the years 2015 to 2018
df_2015 = pd.read_csv(filepath_2015, encoding = 'unicode_escape')
df_2016 = pd.read_csv(filepath_2016, encoding = 'unicode_escape')
df_2017 = pd.read_csv(filepath_2017, encoding = 'unicode_escape')
df_2018 = pd.read_csv(filepath_2018, encoding = 'unicode_escape')

# Combine data from all worksheets as single DataFrame
# df_all = pd.concat(df_dict.values(), ignore_index=True)
```

```python
df_T2014.head()
```

```python
# Extracting all the countries in certain UN_regions and creating seperate lists

asia_list = df_T2014.query("UN_region=='Asia'")["country"]
europe_list = df_T2014.query("UN_region=='Europe'")["country"]
africa_list = df_T2014.query("UN_region=='Africa'")["country"]
latinam_list = df_T2014.query("UN_region=='Latin America and the Caribbean'")["country"]
oceania_list = df_T2014.query("UN_region=='Oceania'")["country"]
northam_list = df_T2014.query("UN_region=='North America'")["country"]
world_list = df_T2014.query("UN_region=='World'")["country"]

print("Asia", len(set(asia_list)))
print("Europe", len(set(europe_list)))
print("Africa", len(set(africa_list)))
print("Latin America", len(set(latinam_list)))
print("Oceania", len(set(oceania_list)))
print("North America", len(set(northam_list)))
print("World", len(set(world_list)))
```

```python
df_T2014 = df_T2014.drop(['UN_subregion'], axis = 1)
df_2015 = df_2015.drop(['Qscore'], axis = 1)
df_2016 = df_2016.drop(['Qscore'], axis = 1)
df_2017 = df_2017.drop(['Qscore'], axis = 1)
df_2018 = df_2018.drop(['Qscore'], axis = 1)

# df_2015.head()
# df_2016.head()
# df_2017.head()
df_2018.head()
```

```python
df_T2014.record.unique()
```

```python
# Rename and re-arrange columns
df_2017.rename({'value': 'total'}, axis=1, inplace=True)
df_T2014.rename({'country_code': 'countryCode', 'crop_land': 'cropLand','fishing_ground': 'fishingGround', 'grazing_land': 'grazingLand', 'forest_land': 'forestLand', 'built_up_land': 'builtupLand', 'value': 'total', 'UN_region': 'UNRegion'}, axis=1, inplace=True)
column_titles = ['year','country','countryCode','record','cropLand', 'grazingLand', 'forestLand', 'fishingGround', 'builtupLand', 'carbon', 'population', 'QScore', 'total']
df_T2016 = df_T2014.reindex(columns=column_titles)
df_2016 = df_2016.reindex(columns=column_titles)
df_2017 = df_2017.reindex(columns=column_titles)
df_2018 = df_2018.reindex(columns=column_titles)
```

```python
# Extracting only the required record types
df_2015_clean = df_2015.query("record == 'BiocapPerCap' | record == 'BiocapTotGHA' | record == 'EFConsPerCap' | record == 'EFConsTotGHA' | record == 'EFExportsPerCap' | record == 'EFExportsTotGHA' | record == 'EFImportsPerCap' | record == 'EFImportsTotGHA' | record == 'EFProdPerCap' | record == 'EFProdTotGHA'")
df_2016_clean = df_2016.query("record == 'BiocapPerCap' | record == 'BiocapTotGHA' | record == 'EFConsPerCap' | record == 'EFConsTotGHA' | record == 'EFExportsPerCap' | record == 'EFExportsTotGHA' | record == 'EFImportsPerCap' | record == 'EFImportsTotGHA' | record == 'EFProdPerCap' | record == 'EFProdTotGHA'")
df_2017_clean = df_2017.query("record == 'BiocapPerCap' | record == 'BiocapTotGHA' | record == 'EFConsPerCap' | record == 'EFConsTotGHA' | record == 'EFExportsPerCap' | record == 'EFExportsTotGHA' | record == 'EFImportsPerCap' | record == 'EFImportsTotGHA' | record == 'EFProdPerCap' | record == 'EFProdTotGHA'")
df_2018_clean = df_2018.query("record == 'BiocapPerCap' | record == 'BiocapTotGHA' | record == 'EFConsPerCap' | record == 'EFConsTotGHA' | record == 'EFExportsPerCap' | record == 'EFExportsTotGHA' | record == 'EFImportsPerCap' | record == 'EFImportsTotGHA' | record == 'EFProdPerCap' | record == 'EFProdTotGHA'")
```

```python
# Combining data from kaggle dataset and data extracted through API
# df_combined now has data from 1961 to 2018
df_combined = pd.concat([df_T2014, df_2015_clean, df_2016_clean, df_2017_clean, df_2018_clean], ignore_index=True)
df_combined = df_combined.drop(['ISO alpha-3 code', 'countryCode', 'QScore'], axis = 1)
df_combined = df_combined.sort_values(by=['year'])

# Saving to CSV
# df_combined.to_csv('Data/Combined.csv')

df_combined.head()
```

### 2. df_combined Dataset Preparation

The following are achieved - 
 - Create seperate lists for each UN Region
 - Append UN Region to data from 2015 to 2018
 - Append 3 digit Country Code to data from 1961 to 2014
 - Creating datasets for countries only and regions only

```python
# Defining filepath
filepath_combined = 'Data/Combined.csv'
filepath_T2014 = 'Data/Till2014.csv'

# Load file with `sheet_name=None` - returns a dictionary
# df = pd.read_csv(filepath_combined, encoding = 'unicode_escape')
df_T2014 = pd.read_csv(filepath_T2014, encoding = 'unicode_escape')
```

```python
# Mapping countries to their UN regions because countries from 2015 do not have UN Regions
for index, row in df_combined.iterrows():
    if df_combined.at[index, 'country'] in set(asia_list):
        df_combined.at[index, 'UNRegion'] = 'Asia'
    elif df_combined.at[index, 'country'] in set(europe_list):
        df_combined.at[index, 'UNRegion'] = 'Europe'
    elif df_combined.at[index, 'country'] in set(africa_list):
        df_combined.at[index, 'UNRegion'] = 'Africa'
    elif df_combined.at[index, 'country'] in set(latinam_list):
        df_combined.at[index, 'UNRegion'] = 'LatinAmerica'
    elif df_combined.at[index, 'country'] in set(oceania_list):
        df_combined.at[index, 'UNRegion'] = 'Oceania'
    elif df_combined.at[index, 'country'] in set(northam_list):
        df_combined.at[index, 'UNRegion'] = 'NorthAmerica'
    elif df_combined.at[index, 'country'] in set(world_list):
        df_combined.at[index, 'UNRegion'] = 'World'
```

```python
countries = list(df_2015.country.unique())
country_code_mapping = {}

for i in countries:
    if i not in country_code_mapping:
        country_code_mapping[i] = []
        country_code_mapping[i].append(df_2015.query(f'country == "{i}"')['countryCode'].unique()[0])

# country_code_mapping

for index, row in df_combined.iterrows():
    if df_combined.at[index, 'country'] in list(country_code_mapping):
        df_combined.at[index, 'countryCode'] = country_code_mapping[f'{row.country}'][0]
```

```python
big_regions = ['Africa', 'Asia', 'Europe', 'Latin America and the Caribbean', 'North America', 'Oceania', 
               'Australia and New Zealand', 'Caribbean', 'Central America', 'Central Asia', 
              'Eastern Africa', 'Eastern Asia', 'Eastern Europe', 'Melanesia', 'Middle Africa', 'Northern Africa', 
               'Northern Europe', 'Polynesia', 'South-Eastern Asia', 'South America', 'Southern Africa',
               'Southern Asia', 'Southern Europe', 'Western Africa', 'Western Asia', 'Western Europe']

df_countries =  df_combined[~df_combined['country'].isin(big_regions)]

df_regions = df_combined[df_combined['country'].isin(big_regions)]

len(set(df_countries.country))
```
